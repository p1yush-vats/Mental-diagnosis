# Encoding Features

```python
{column: len(data[column].unique()) for column in data.select_dtypes('object').columns}
```

```python
{column: list(data[column].unique()) for column in data.select_dtypes('object').columns}
```

```python
def clean_gender(g):
    g = str(g).strip().lower()
    if g in ['male', 'm', 'man', 'cis male', 'male ', 'cis man']:
        return 'Male'
    elif g in ['female', 'f', 'woman', 'cis female', 'female ']:
        return 'Female'
    else:
        return 'Other'
```

```python
data['Gender'] = data['Gender'].apply(clean_gender)
```

```python
target = 'treatment'

binary_features = [
    'self_employed',
    'family_history',
    'remote_work',
    'tech_company',
    'obs_consequence'
]

ordinal_features = [
    'work_interfere',
    'no_employees',
    'leave'
]

nominal_features = [
    'Gender',
    'Country',
    'state',
    'benefits',
    'care_options',
    'wellness_program',
    'seek_help',
    'anonymity',
    'mental_health_consequence',
    'phys_health_consequence',
    'coworkers',
    'supervisor',
    'mental_health_interview',
    'phys_health_interview',
    'mental_vs_physical'
]
```

```python
def binary_encode(df, columns, positive_values):
    df = df.copy()
    for column, positive_value in zip(columns, positive_values):
        df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0)
    return df



def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data['leave'].unique()
```

```python
binary_positive_values = ['Yes' for feature in binary_features]



ordinal_mappings = {
    'work_interfere': {
        'Never': 0,
        'Rarely': 1,
        'Sometimes': 2,
        'Often': 3
    },
    'no_employees': {
        '1-5': 0,
        '6-25': 1,
        '26-100': 2,
        '100-500': 3,
        '500-1000': 4,
        'More than 1000': 5
    },
    'leave': {
        'Very easy': 4,
        'Somewhat easy': 3,
        "Don't know": 2,
        'Somewhat difficult': 1,
        'Very difficult': 0
    }
}


nominal_prefixes = [
    'ge',
    'co',
    'st',
    'be',
    'ca',
    'we',
    'se',
    'an',
    'mc',
    'ph',
    'cw',
    'su',
    'mi',
    'pi',
    'mp'
]
```

```python
# --- Binary Encoding ---
for col in binary_features:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])


# --- ordinal mapping
for col, mapping in ordinal_mappings.items():
    data[col] = data[col].map(mapping)


# --- One-Hot Encoding (Nominal) ---
ohe = OneHotEncoder(drop='first', sparse_output=False)  # ✅ For older versions (<1.2)

nominal_encoded = ohe.fit_transform(data[nominal_features])

# Get column names and create DataFrame
ohe_cols = ohe.get_feature_names_out(nominal_features)
nominal_df = pd.DataFrame(nominal_encoded, columns=ohe_cols, index=data.index)

# Drop original nominal columns and concat new one-hot encoded columns
data = pd.concat([data.drop(columns=nominal_features), nominal_df], axis=1)
```

```python
final_feature_names = binary_features + list(ordinal_mappings.keys()) + list(ohe_cols)
data
```

```python
y = data['treatment'].copy()
X = data.drop('treatment', axis=1).copy()
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

# Results

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_auc_score, classification_report, RocCurveDisplay
)
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Initialize models
log_reg = LogisticRegression(max_iter=1000, random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)

models = {
    "Logistic Regression": log_reg,
    "Random Forest": rf
}

# Train, predict, evaluate
for name, model in models.items():
    print(f"\n--- {name} ---")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # for ROC AUC

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    # Plot ROC Curve
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.title(f"ROC Curve: {name}")
    plt.show()

feature_names = final_feature_names
importances = rf.feature_importances_

# Now both lists will be of the same length
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

top_10 = importance_df.sort_values(by='Importance', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=top_10, palette='mako', hue='Feature')
plt.title('Top 10 Important Features (Random Forest)')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```
