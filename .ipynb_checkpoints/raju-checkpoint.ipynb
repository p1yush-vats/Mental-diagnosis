# Encoding Features

```python
{column: len(data[column].unique()) for column in data.select_dtypes('object').columns}
```

```python
{column: list(data[column].unique()) for column in data.select_dtypes('object').columns}
```

```python
def clean_gender(g):
    g = str(g).strip().lower()
    if g in ['male', 'm', 'man', 'cis male', 'male ', 'cis man']:
        return 'Male'
    elif g in ['female', 'f', 'woman', 'cis female', 'female ']:
        return 'Female'
    else:
        return 'Other'
```

```python
data['Gender'] = data['Gender'].apply(clean_gender)
```

```python
target = 'treatment'

binary_features = [
    'self_employed',
    'family_history',
    'remote_work',
    'tech_company',
    'obs_consequence'
]

ordinal_features = [
    'work_interfere',
    'no_employees',
    'leave'
]

nominal_features = [
    'Gender',
    'Country',
    'state',
    'benefits',
    'care_options',
    'wellness_program',
    'seek_help',
    'anonymity',
    'mental_health_consequence',
    'phys_health_consequence',
    'coworkers',
    'supervisor',
    'mental_health_interview',
    'phys_health_interview',
    'mental_vs_physical'
]
```

```python
def binary_encode(df, columns, positive_values):
    df = df.copy()
    for column, positive_value in zip(columns, positive_values):
        df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0)
    return df



def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data['leave'].unique()
```

```python
binary_positive_values = ['Yes' for feature in binary_features]



ordinal_mappings = {
    'work_interfere': {
        'Never': 0,
        'Rarely': 1,
        'Sometimes': 2,
        'Often': 3
    },
    'no_employees': {
        '1-5': 0,
        '6-25': 1,
        '26-100': 2,
        '100-500': 3,
        '500-1000': 4,
        'More than 1000': 5
    },
    'leave': {
        'Very easy': 4,
        'Somewhat easy': 3,
        "Don't know": 2,
        'Somewhat difficult': 1,
        'Very difficult': 0
    }
}


nominal_prefixes = [
    'ge',
    'co',
    'st',
    'be',
    'ca',
    'we',
    'se',
    'an',
    'mc',
    'ph',
    'cw',
    'su',
    'mi',
    'pi',
    'mp'
]
```

```python
# --- Binary Encoding ---
for col in binary_features:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])


# --- nominal mapping
for col, mapping in ordinal_mappings.items():
    data[col] = data[col].map(mapping)


# --- One-Hot Encoding (Nominal) ---
ohe = OneHotEncoder(drop='first', sparse=False)  # ✅ For older versions (<1.2)

nominal_encoded = ohe.fit_transform(data[nominal_features])

# Get column names and create DataFrame
ohe_cols = ohe.get_feature_names_out(nominal_features)
nominal_df = pd.DataFrame(nominal_encoded, columns=ohe_cols, index=data.index)

# Drop original nominal columns and concat new one-hot encoded columns
data = pd.concat([data.drop(columns=nominal_features), nominal_df], axis=1)
```

```python
data
```

# Encoding Labels

```python
data = binary_encode(data, columns=['treatment'], positive_values=['Yes'])
```

```python
print("Remaining non-numeric columns:", len(data.select_dtypes('object').columns))
```

```python
print("Remaining missing values:", data.isna().sum().sum())
```

# Splitting/Scaling

```python
y = data['treatment'].copy()
X = data.drop('treatment', axis=1).copy()
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

# Training

```python
X.shape
```

```python
print("Class Distribution (Positive to Negative): {:.1f}% / {:.1f}%".format(y_train.mean() * 100, (1 - y_train.mean()) * 100))
```

```python
inputs = tf.keras.Input(shape=(X.shape[1],))
x = tf.keras.layers.Dense(1024, activation='relu')(inputs)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs, outputs)


model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ]
)


batch_size = 64
epochs = 50

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau()
    ]
)
```

# Results

```python
plt.figure(figsize=(12, 6))

plt.plot(range(epochs), history.history['accuracy'], label="Training Accuracy")
plt.plot(range(epochs), history.history['val_accuracy'], label="Validation Accuracy")

plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.legend()
plt.title("Accuracy Over Time")

plt.show()
```

```python
model.evaluate(X_test, y_test)
```
